```python
from bqplot import OrdinalScale, LinearScale, Bars, Figure, Axis, ColorScale, ColorAxis, Tooltip

def make_bar(x_data, y_data, colormap, xaxis_label, xaxis_tick_size, yaxis_label, xhover_label, yhover_label, fig_title):
    x_data = x_data
    y_data = y_data
    x_scale = OrdinalScale()
    y_scale = LinearScale()
    col_sc = ColorScale(scheme = colormap)
    ax_x = Axis(scale=x_scale, grid_lines = 'none', label = xaxis_label, label_location='middle', label_offset='40px', tick_rotate=-15, tick_style = {'font-size': xaxis_tick_size}) # e.g. "8px"
    ax_y = Axis(scale=y_scale, orientation='vertical', grid_lines='none', label=yaxis_label, label_offset='50px')
    ax_c = ColorAxis(scale=col_sc, side = 'right', num_ticks = 2)
    tooltip = Tooltip(fields=["x", "y"], labels=[xhover_label, yhover_label], formats=["", ".1%"])
    # 'i' as a format refers to integer
    bar = Bars(x=x_data, y=y_data, color=y_data, scales={'x': x_scale, 'y':y_scale, 'color': col_sc}, tooltip = tooltip, label_display=True,
                  label_font_style={'font-size':'10px', 'fill':'white', 'font-weight':'bold'}, label_display_format='.1%',
                      interactions={"click": "select", "hover": "tooltip"}, unselected_style = {"opacity": 0.3}, selected_style={"stroke": "magenta", "stroke-width": "2"})
    fig = Figure(title = fig_title, marks=[bar], axes=[ax_x, ax_y, ax_c])
    return fig
```

```python
def age_today(i):
    today=date.today()
    raw_years_to_adjust = today.year - i.year
    if (today.month, today.day) < (i.month, i.day) == True: # If today's date is earlier in the year than the birthdate i.e. pending birthday
        raw_years_to_adjust -= 1
    elif (i.month, i.day) < (today.month, today.day) == True: # If the birthdate is earlier in the year than today's date i.e. birthday already passed
        raw_years_to_adjust -= 0
    age = raw_years_to_adjust
    return age
```
